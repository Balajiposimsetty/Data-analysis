```python
#plot the scores of two students aginst 5 subjects.subjects as x-axis,marks as y-axis
s1=[56,78,92,95,78]
s2=[77,89,91,23,45]
sub=["mat","sci","eng","tel","soc"]
plt.plot(sub,s1,label="student1",color="b")
plt.plot(sub,s2,label="student2",color="y")
plt.title("Marks of student")
plt.legend()
```

```python
s1=[56,78,92,95,78]
s2=[77,89,91,23,45]
sub=["mat","sci","eng","tel","soc"]
plt.subplot(2,2,1)
plt.plot(sub,s1,label="student1",color="b")
plt.legend()
plt.subplot(2,2,2)
plt.plot(sub,s2,label="student2",color="g")
plt.legend()
plt.subplot(2,2,3)
plt.plot(sub,s2,label="student2",color="y")
plt.legend()
plt.subplot(2,2,4)
plt.plot(sub,s1,label="student1",color="r")
plt.legend()
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenues of company A",A)
print("Revenues of company B",B)
proA=np.diff(A)
proB=np.diff(B)
print("Profits of comany A",proA)
print("Profits of comany B",proB)
year=["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(year,proA,label="Profit-A",color="k")
plt.legend(loc= "upper right")
plt.subplot(1,2,2)
plt.bar(year,proB,label="Profit-B",color="g")
plt.legend(loc= "upper left")
```
